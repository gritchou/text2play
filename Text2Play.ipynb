```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Image

def search_wiki_pages(search_term):
    url = f"https://en.wikipedia.org/w/api.php?action=query&list=search&format=json&srsearch={search_term}"
    response = requests.get(url)
    data = response.json()
    search_results = [result['title'] for result in data['query']['search']]
    return search_results

def get_wiki_page_content(page_title):
    url = f"https://en.wikipedia.org/w/api.php?action=query&titles={page_title}&prop=extracts|images&format=json"
    response = requests.get(url)
    data = response.json()
    page_id = list(data['query']['pages'].keys())[0]
    page_content = data['query']['pages'][page_id]['extract']
    soup = BeautifulSoup(page_content, 'html.parser')
    page_text = soup.get_text()
    image_url = None
    if 'images' in data['query']['pages'][page_id]:
        for image_info in data['query']['pages'][page_id]['images']:
            if 'File:' in image_info['title']:
                image_title = image_info['title']
                image_url = f"https://en.wikipedia.org/wiki/{image_title.replace(' ', '_')}"
                break

    return page_text, image_url

def get_painting_url(page_title):
    response = requests.get(image_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    div_img=soup.find('div', {'id':'file'})
    img_tag = div_img.find('img')
    img_src = img_tag.get('src')
    return img_src

search_results = search_wiki_pages('Mona Lisa Leonardo da Vinci')[0]
content_results, image_url = get_wiki_page_content(search_results)
print(content_results)
print(image_url)
img_src=get_painting_url(image_url)
print(img_src)
list_words=content_results.split(' ')
words = [word.replace('(', '') for word in list_words]
print(words)
```
